```python
#Next level itemset generation function
def nextgen(A):
    N=[]
    for l in range(0,len(A)): 
        t1=list(A)[l]
        sub = list(itertools.combinations(t1,len(t1)-1))
        for k in range(l+2,len(A)):
            t2=list(A)[k]
            S2 = set(t2)
            for i in range(0,len(sub)):
                S1 = set(sub[i])
                if S1.issubset(S2):
                    ts=set(t1).union(S2)
                    N.append(tuple(ts))
    return N
        
```

```python
#MSCandidate-Gen Function 
#Input: Fk-1, SDC
#Output: Ck, k>=3
def CandidateGen(C,SDC):
    Ck=[]
    Combined = nextgen(C)
    for i in range(0,len(Combined)):
        sup = list(itertools.combinations(Combined[i],2))
        #diff=abs(S.get(sup[j][0])-S.get(sup[j][1]))
        for j in range(0,len(sup)):
            diff=abs(S.get(sup[j][0])-S.get(sup[j][1]))
            if diff>SDC:
                break
        if j==len(sup)-1:
            Ck.append(Combined[i])
    #print(Ck)
    return Ck
                
```
